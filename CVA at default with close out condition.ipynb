# Abstruct

We consider two type of the close-out out condition, par and risky. 
Legal documents such as ISDA allow close-out value take into account bid-offer at the default\[1\]).
At the counterparty's default, close-out value can include xVA.
In the paper, we fucus on the close out value with or without unilateral CVA and see the CVA difference.
Under the risky close-out, we conculded close-out values modulated $V(\tau) \to  V(\tau) \exp[1-R]$.

# Formulation

## Intoroduction of PDE

Total ortofolio value including xVA $\hat{V}$  can be described as following PDE(\[1\]);

$$
(\partial_t + \mathcal{A}_t -r) \hat{V} = (1 -R) \lambda M^+ 
$$
$$
\hat{V}(T, S) = H( S)
$$

We ignore the the other xVA but CVA.
The PDE can be decomposed into defalut risk free valus $V$ and CVA $U$, as

$$
(\partial_t + \mathcal{A}_t -r) V = 0 
$$
$$
V(T, S) = H( S)
$$

$$
(\partial_t + \mathcal{A}_t -r) U = (1 -R) \lambda M^+ 
$$
$$
U(T) = 0
$$

| notation | description |
| :------- | :---------- |
| $V$ | risk-free potfolio value |
| $\hat{V}$ | risky potfolio value, $\hat{V}= V +U$  |
| $U$ | Unilateral CVA |
| $M$ | Close-out value at counterparty default |
| $r$ | risk-free rate |
| $\lambda$ | credit spread, from bond yield ($r_c - r$) |
| $R$ | recovery rate, set 40%|
| $\mathcal{A}_t$ | differencial operator |
| $S$ | derivative underlying |
| $x$ | $\log S$|
| $\mu$ | drift of underlying $S$ |
| $\sigma$ | volativity of underlying $S$ |
| $H(x)$ | risk-free portfolio value at maturity with underlyng $x$|

for instance, we introduce the parabolic differential operator$\mathcal{A}_t$

$$
\mathcal{A}_t \equiv \frac{1}{2} \sigma^2S^2 \partial_S^2 + \mu S \partial_S
$$
or
$$
\mathcal{A}_t=\frac{1}{2} \sigma^2\partial_x^2 + \mu \partial_x
$$

let us consider two type of close-out condition; par close-out $M = V$, and, risky close-out $M = \hat{V} = V +U$.
We obtain two PDE for CVA $U$ for each conditions:

- Close-out value $M = V$
$$
(\partial_t + \mathcal{A}_t -r) U = (1 -R) \lambda V^+ 
$$
$$
U(T) = 0
$$
- Close-out value $M = \hat{V}$
$$
(\partial_t + \mathcal{A}_t -r) U = (1 -R) \lambda (V + U)^+ 
$$
$$
U(T) = 0
$$

## Finite differential method

These PDE can not be solved easily. In order to see the behavior, we employ finite diffential approximation and solve them nunerically.
For the calculation, we apply implicit method, which is more stable and easily converged than explicit method. 
Let us denote $V_{i}^{\ j} \equiv V(t_{i}, x_j)$, PDE can be aprpoximated with implicit finite deffenetial method as follows;

$$
(V_{i+1}^{\ j} - V_{i}^{\ j})/\Delta t +\frac{\sigma^2}{2}(V_{i}^{\ j+1} - 2V_{i}^{\ j} + V_{i}^{\ j-1})/\Delta x^2
+\mu(V_{i}^{\ j+1} -  V_{i}^{\ j-1})/2\Delta x - r V_{i}^{\ j} = 0
$$

$$
\therefore V_{i+1}^{\ j}=  
-\left( \frac{\sigma^2 \Delta t }{2 \Delta x^2} + \frac{\mu \Delta t }{2 \Delta x}  \right) V_{i}^{\ j+1}
+ \left(1  +r \Delta t + \frac{\sigma^2  \Delta t }{\Delta x^2}\right)V_{i}^{\ j}
- \left( \frac{\sigma^2 \Delta t }{2 \Delta x^2} - \frac{\mu \Delta t }{2 \Delta x}  \right)V_{i}^{\ j-1}
$$

$$
(U_{i+1}^{\ j} - U_{i}^{\ j})/\Delta t +\frac{\sigma^2  }{2}(U_{i}^{\ j+1} - 2U_{i}^{\ j} + U_{i}^{\ j-1})/\Delta x^2
+\mu(V_{i}^{\ j+1} -  V_{i}^{\ j-1})/2\Delta x -r U_{i}^{\ j}= (1 -R) \lambda (M_{i}^{\ j})^+
$$

$$
\therefore U_{i+1}^{\ j} = 
-\left( \frac{\sigma^2 \Delta t }{2 \Delta x^2} + \frac{\mu \Delta t }{2 \Delta x}  \right) U_{i}^{\ j+1}
+ \left(1  +r \Delta t + \frac{\sigma^2  \Delta t }{\Delta x^2}\right)U_{i}^{\ j}
- \left( \frac{\sigma^2 \Delta t }{2 \Delta x^2} - \frac{\mu \Delta t }{2 \Delta x}  \right)U_{i}^{\ j-1}
+ (1 - R) \lambda (M_{i}^{\ j})^+ \Delta t
$$

with 
$$
V_I^j = H^j, U_I^j= 0
$$

For simplicity, close-out values are evaluated in future time $(i+1)$, i.e., replacing $M_{i}^{\ j}$ from finite differential equation into $M_{i+1}^{\ j}$

$$
U_{i+1}^{\ j} = 
-\left( \frac{\sigma^2 \Delta t }{2 \Delta x^2} + \frac{\mu \Delta t }{2 \Delta x}  \right) U_{i}^{\ j+1}
+ \left(1  +r \Delta t + \frac{\sigma^2  \Delta t }{\Delta x^2}\right)U_{i}^{\ j}
- \left( \frac{\sigma^2 \Delta t }{2 \Delta x^2} - \frac{\mu \Delta t }{2 \Delta x}  \right)U_{i}^{\ j-1}
+ (1 - R) \lambda (M_{i+1}^{\ j})^+ \Delta t
$$

## Close out value $M_i = V_{i+1} + U_{i+1}$

It is intuitive because close-out value is reduced by CVA .

With the risky close-out, CVA no more converges LGD * PV, but LGD' * PV.

```python
lgd_mod = 1.0 - np.exp(-(1.0-recovery))
plt.xlim([90,110])
plt.ylim([-10,10])
view_time = 0
plt.plot(fdm_good.space_grid, (1.0-recovery) * fdm_good.pv_grid[view_time], label='LGD * PV')
plt.plot(fdm_good.space_grid, lgd_mod * fdm_good.pv_grid[view_time], color='green', label='LGD\' * PV')
plt.plot(fdm_good.space_grid, -fdm_bad.cva_grid(True)[view_time], label='- CVA(bad, risky)', color='purple',  linestyle='--')
plt.plot(fdm_good.space_grid, np.zeros(len(fdm_good.space_grid)), color='black',  linestyle=':')
plt.legend();
```

where, $LGD' = 1 - \exp[1-R]$, or $R' = \exp[-(1-R)] $. Risky close-out modulates recovery rate $R$ into $R'$.
Under $V(\tau) > 0$, close out value at defalut $M$ is,
$$
M(\tau) = V(\tau) + U(\tau) = V(\tau) - (1 - \exp[1-R]) V(\tau) =  \exp[1-R] V(\tau)
$$

Farthermore, total loss $L(\tau)$ for bank compensated by CVA differs by close-out.
$$
L(\tau) = 
\begin{cases}
(1-R) M(\tau) + U(\tau) = 0 & \text{for par close-out}\\
(1-R) M(\tau) = (1-R) e^{1-R}V(\tau)& \text{for risy close-out}\\
\end{cases}
$$
If risky close-out appliesd, default risk is no more hedged.

# Conclusion

| items| par close-out $M=V$&nbsp; &nbsp;| risky close-out $M=\hat{V}$ &nbsp; &nbsp;|
| :-------------: | :-------------------: | :-------------------:|
| CVA $U$ | larger | smaller |
| CVA at default $U(\tau)$ | $-(1-R) V(\tau)$  | $-(1-e^{1-R}) V(\tau)$  |
| Close-out  $M(\tau)$ | $V(\tau)$  | $e^{1-R} V(\tau)$  |
| total loss  $L(\tau)$ | $0$  | $(1-R)e^{1-R} V(\tau)$  |

# Appendix; Implimation

```python
%matplotlib inline
import numpy as np
from scipy import statss
import pandas as pd
from scipy import sparse 
import matplotlib as mpl
import matplotlib.pyplot as plt
import subprocess
import seaborn as sns
```

```python
class ImplicitFdmCva:
    def __init__(self, len_x, dx, s0, len_t, dt,t0=0.0):
        x0 = np.log(s0)
        self._dt = dx
        self._dx = dt
        self._grid_x = np.arange(x0 - 0.5 * len_x * dx, x0 + 0.5 * len_x * dx, dx)
        self._grid_s = np.exp(self._grid_x)
        self._grid_t = np.arange(0.0, len_t * dt, dt)
        self._arr_pv = [np.zeros(len_x) for i in range(len_t)]
        self._arr_cva = [np.zeros(len_x) for i in range(len_t)]
        self._arr_cva_risky = [np.zeros(len_x) for i in range(len_t)]
        return    
    def calc_pv_grid(self, drift, volatility, rfr, pv_at_maturity):
        self._arr_pv[-1] = pv_at_maturity(np.exp(self._grid_x))
        self._build_pv_pde_matrix(drift, volatility, rfr)
        len_time = len(self._grid_t)
        delta_condition = self._grid_s[-1] - self._grid_s[-2]
        for i in np.arange(1, len_time, 1):
            arr_pv_border = self._arr_pv[len_time - i]
            arr_pv_border[0] = 0.0
            arr_pv_border[-1] = delta_condition
            self._arr_pv[len_time -1- i] =  np.dot(self._mat_pv_pde, arr_pv_border)
        return

    def _build_pv_pde_matrix(self, drift, volatility, rfr):
        len_space = len(self._grid_x)
        arr_diag = (1.0  +  rfr * self._dt + volatility**2 * self._dt / self._dx**2 ) * np.ones(len_space)
        arr_diag_plus = -0.5 * self._dt * (volatility**2 / self._dx**2 + drift/self._dx) * np.ones(len_space)
        arr_diag_minus = -0.5 * self._dt * (volatility**2 / self._dx**2 - drift/self._dx) * np.ones(len_space)
        mat_imp_fwd_driver = sparse.dia_matrix(\
           ([arr_diag_plus, arr_diag, arr_diag_minus],  np.array([-1,0,1])), \
            shape=[len_space, len_space])
        mat_imp_fwd_driver.data[1][0] = 1.0
        mat_imp_fwd_driver.data[2][1] = -1.0
        mat_imp_fwd_driver.data[1][-1] = 1.0
        mat_imp_fwd_driver.data[0][-2] = -1.0
        self._mat_pv_pde = np.linalg.inv(mat_imp_fwd_driver * np.identity(len_space))
        return
        
    def calc_cva_grid(self, recovery, est_default_time):
        self._build_default_curve(est_default_time)
        self._recovery = recovery
        self._arr_cva = self._build_cva_grid(isRisky=False)
        self._arr_cva_risky = self._build_cva_grid(isRisky=True)
        return
    
    def _build_cva_grid(self, isRisky=False):
        len_space = len(self._grid_x)
        len_time = len(self._grid_t)
        arr_cva = [np.zeros(len_space) for i in range(len_time)]
        
        def close_out(time, isRisky):
            arr_full_pv = (self._arr_pv[time]  + arr_cva[time]) if isRisky else self._arr_pv[time]
            return 0.5 * (arr_full_pv + abs(arr_full_pv))
        
        for i in np.arange(1, len_time, 1):
            delta_condition = (arr_cva[len_time - i][-1]  - arr_cva[len_time - i][-2])  * (i+1) / i
            arr_driven_border = arr_cva[len_time - i] \
                - (1.0 - self._recovery) * self._hazard[len_time - i] * self._dt * close_out(len_time - i, isRisky)
            arr_driven_border[0] = 0.0
            arr_driven_border[-1] = delta_condition
            arr_cva[len_time - 1 - i] =  np.dot(self._mat_pv_pde, arr_driven_border)
        return arr_cva
    
    def _build_default_curve(self, default_time):
        self._hazard = stats.expon.pdf(self._grid_t, loc=0, scale=default_time)
    
    def get_pv(self, s):
        idx_s = self._grid_s.searchsorted(s)
        return [self._arr_pv[0][idx_s], self._grid_s[idx_s]]
    
    def get_cva(self, s, isRisky=False):
        idx_s = self._grid_s.searchsorted(s)
        cva = self._arr_cva_risky[0][idx_s] if isRisky else self._arr_cva[0][idx_s]
        return [cva, self._grid_s[idx_s]]
    
    @property
    def pv_grid(self):
        return self._arr_pv
    @property
    def time_grid(self):
        return  self._grid_t
    @property
    def space_grid(self):
        return  self._grid_s
    @property
    def hazard(self):
        return self._hazard
    def cva_grid(self, is_risky = False):
        return self._arr_cva_risky if is_risky else self._arr_cva
```
